```python
# Seleccionar solo las columnas numéricas
numeric_columns = df.select_dtypes(include=[np.number])

# Calcula la matriz de correlación en las columnas numéricas
correlation_matrix = numeric_columns.corr()

# Mostrar la matriz de correlación y el mapa de calor
sns.heatmap(correlation_matrix, annot=True)
plt.title("Matriz de correlación para características numéricas")
plt.xlabel("Características de la película")
plt.ylabel("Movie features")
plt.show()
```

```python
# Calcula una matriz de correlación para las columnas del DataFrame después de convertir las etiquetas categóricas en valores numéricos.
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

# Desglosa la matriz de correlación en pares de correlación.
corr_pairs = correlation_mat.unstack()

# Imprime los pares de correlación.
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# visualizar las que tienen una alta correlación (> 0,5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# agrupar el df por las columnas 'company' y 'year' y calcular la suma de la columna 'gross' para cada grupo.

df.groupby(['company', 'year'])[["gross"]].sum()
```

```python
# Calcular la suma de los ingresos brutos ('gross') para cada compañía en cada año.
CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

# Ordena la tabla resumen en orden descendente según la columna 'gross' y, en caso de empate, por 'company' y 'year'.
# Luego, selecciona las 20 filas principales con los mayores ingresos brutos.
CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross', 'company', 'year'], ascending=False)[:20]

# Convierte la columna 'gross' en números enteros (int64) para una representación más adecuada de los ingresos.
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64')

# Muestra la lista de las 20 compañías con los mayores ingresos brutos.
CompanyGrossSumSorted
```

```python
# Calcula la suma de los ingresos brutos ('gross') para cada compañía en el DataFrame, agrupado por compañía.
CompanyGrossSum = df.groupby(['company'])[["gross"]].sum()

# Ordena la tabla de resumen en orden descendente según la columna 'gross' y, en caso de empate, por el nombre de la compañía 'company'.
# Luego, selecciona las 20 compañías principales con los mayores ingresos brutos.
CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross', 'company'], ascending=False)[:20]

# Convierte la columna 'gross' en números enteros (int64) para una representación más adecuada de los ingresos.
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64')

# Muestra la lista de las 20 compañías con los mayores ingresos brutos.
CompanyGrossSumSorted
```

```python
# Crea un gráfico de dispersión (scatter plot) con el presupuesto (budget) en el eje x y las ganancias brutas (gross) en el eje y.
# El parámetro alpha=0.5 controla la transparencia de los puntos en el gráfico, lo que puede ayudar a ver la densidad de los datos.

plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Presupuesto vs Ingresos brutos')
plt.xlabel('Ingresos brutos')
plt.ylabel('Presupuesto por película')
plt.show()
```

```python
# El DataFrame df_numerized ahora contiene las mismas columnas, pero las columnas categóricas se han convertidas en numéricas.

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```
